```python
import numpy as np
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
from matplotlib import pyplot
import seaborn as sns
import scipy.stats  as stats
```

```python
data = pd.read_csv('eth.csv', index_col=False)
pd.options.display.float_format = '{:,.2f}'.format
```

```python
NaN_rows = data[data.isnull().any(1)]
new_data = data[~data.index.isin(NaN_rows.index)]
new_data.nunique()
#data.dropna()

new_data['date'] = pd.to_datetime(new_data['date'])
```

```python
a4_dims = (10, 8)
fig, ax = pyplot.subplots(figsize=a4_dims)
var_corr = new_data.corr()

sns.heatmap(var_corr, xticklabels=var_corr.columns, yticklabels=var_corr.columns, annot=True)
#seems like all variables have negative correlation with generated coins
```

```python
new_data2 = new_data.drop(['medianFee','blockCount'],1)
a4_dims = (10, 8)
fig, ax = pyplot.subplots(figsize=a4_dims)
var_corr2 = new_data2.corr()

sns.heatmap(var_corr2, xticklabels=var_corr2.columns, yticklabels=var_corr2.columns, annot=True)
#seems like all variables have negative correlation with generated coins
```

```python
var_corr3 = var_corr2[(abs(var_corr2) > 0.6) & (abs(var_corr2) < 1)]
a4_dims = (10, 8)
fig, ax = pyplot.subplots(figsize=a4_dims)
sns.heatmap(var_corr3, xticklabels=var_corr3.columns, yticklabels=var_corr3.columns, annot=True)
```

```python
new_data3 = new_data2.drop(['medianTxValue(USD)'],1)
var_corr4 = new_data3.corr()

a4_dims = (10, 8)
fig, ax = pyplot.subplots(figsize=a4_dims)

heatmap = sns.heatmap(var_corr4, xticklabels=var_corr4.columns, yticklabels=var_corr4.columns, annot=True)
heatmap.figure.savefig('heatmap.png')
```
